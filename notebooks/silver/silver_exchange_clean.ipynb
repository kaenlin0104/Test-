```python
# ============================================
# SILVER LAYER - EXCHANGE CLEANING
# ============================================

import logging
from pyspark.sql.functions import col
from pyspark.sql.types import IntegerType, DoubleType
from utils.validation import validate_exchange

logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

# LOAD
df = spark.read.parquet("Files/Bronze/processed/exchange_raw")

# TRANSFORM
df = df.withColumn("year", col("year").cast(IntegerType())) \
       .withColumn("month", col("month").cast(IntegerType())) \
       .withColumn("exchange_rate", col("exchange_rate").cast(DoubleType()))

# VALIDATE
valid_df, invalid_df = validate_exchange(df)

# WRITE
valid_df.write.mode("overwrite").saveAsTable("silver_exchange_rate_clean")
invalid_df.write.mode("overwrite").saveAsTable("silver_exchange_rate_quarantine")

print("Sample exchange:")
spark.read.table("silver_exchange_rate_clean").limit(5).show()

mssparkutils.notebook.exit("Success")
```
